```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io
from google.colab import files

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

Importing the data


```python
print(sns.get_dataset_names())
```

```python
crash_df = sns.load_dataset('car_crashes')
crash_df
```

DISTRIBUTION PLOTS


1. Distribution plot

```python
crash_df.columns
```

```python
sns.distplot(crash_df['not_distracted'])
```

To remove the Kenel density estimation put kde =false

```python
sns.displot(crash_df['not_distracted'], kde=False)
```

using bins in the dataset to seperate with certain interval

```python
sns.displot(crash_df['not_distracted'], kde=False, bins=20)
```

2. JOINT PLOT

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

Hex showing the distributions

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hex')
```

3. KDE PLOT

```python
sns.kdeplot(crash_df['speeding'])
```

4.PAIR PLOTS

GIVE THE RELATIONSHIP OF ALL THE VARIABLE IN THE DATASET

```python
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
sns.pairplot(tips_df, hue='sex', palette='Blues')
```

5. RUG PLOT

```python
tips_df.columns
```

```python
sns.rugplot(tips_df['total_bill'])
```

STYLING

```python
sns.set_style('ticks')
#plt.figure(figsize=(10,8))

sns.set_context('poster', font_scale=0.5)

#sns.despine(left=True)
sns.countplot(x='sex', data=tips_df)
```

```python
sns.set_style('ticks')
#plt.figure(figsize=(10,8))

sns.set_context('poster', font_scale=0.5)

sns.despine(left=False)
sns.countplot(x='sex', data=tips_df)
```

CATEGORICAL PLOTING

1. BAR PLOTS

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.std)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.mean)
```

2. COUNT PLOT

```python
sns.countplot(x='sex', data=tips_df)
```

3. BOX PLOT

```python
sns.boxplot(x='sex', y='total_bill', data=tips_df, hue='sex')
plt.legend(loc=0)
```

4. VIOLIN

A COMBINATION OF BOX PLOT AND KDE PLOT

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', split=True)
```

6. STRIP PLOT

```python
plt.figure(figsize=(8,5))

sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', hue='sex', data=tips_df, jitter=True)
```

7. SWARM PLOT

IT LIKE THE STRIP PLOT BUT THE POINTS ARE BEING ADJUSTED IN A WAY THAT THERE DONT OVERWRAP EACH OTHER

```python
sns.violinplot(x='day', y='total_bill', data=tips_df)
sns.swarmplot(x='day', y='total_bill', data=tips_df, color='white')
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips_df, color='blue')
```

PALETTES

```python
plt.figure(figsize=(8,5))
sns.set_style('dark')
sns.set_context('talk')
sns.set_palette('rainbow')
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', palette='magma')
```
