```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
crash_df = sns.load_dataset('car_crashes')
crash_df.head(5)
```

```python
crash_df.columns
```

```python
sns.distplot(crash_df['not_distracted'], kde=False, bins=25)
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='kde')
# kde - kenel density estimations
```

```python
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head(5)
```

```python
sns.pairplot(tips_df, hue='sex', palette='Blues')
```

```python
sns.rugplot(tips_df['total_bill'])
```

```python
sns.barplot(x='sex', y='total_bill', data=tips_df)
```

```python
sns.countplot(x='sex', data=tips_df)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex', palette='Reds')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', palette='Oranges', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', jitter=True, dodge=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, palette='Oranges')
sns.swarmplot(x='day', y='total_bill', data=tips_df, dodge=True, palette='Reds')
```

```python
flights = sns.load_dataset('flights')
flights.head(5)
```

```python
flights = flights.pivot_table(index='month', columns='year', values='passengers')
flights
```

```python
sns.heatmap(flights, cmap='Blues', linecolor='White', linewidths=1)
```
