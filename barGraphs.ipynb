```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

# 1. Vertical Bar Chart (Column Chart)
Definition:
Bars are drawn vertically (upwards) — the height of each bar represents the value.

Used for:
Comparing categories or groups across a single variable.

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values = [10, 15, 7]

plt.bar(categories, values)
plt.title("Vertical Bar Chart")
plt.show()
```

#2. Horizontal Bar Chart
*Definition:*
Bars are drawn horizontally, with the length representing the value.

*Used for:*

Long category names (easier to read horizontally).

Emphasizing comparisons between items.

```python
plt.barh(categories, values)
plt.title("Horizontal Bar Chart")
plt.show()
```

#3. Grouped (Clustered) Bar Chart
Definition:
Shows multiple bars for each category, representing different subcategories.

Used for:
Comparing multiple variables across the same categories.

```python
years = ['2023', '2024']
categories = ['A', 'B', 'C']
values_2023 = [10, 15, 7]
values_2024 = [12, 18, 9]

x = np.arange(len(categories))
width = 0.35

plt.bar(x - width/2, values_2023, width, label='2023')
plt.bar(x + width/2, values_2024, width, label='2024')
plt.xticks(x, categories)
plt.legend()
plt.title("Grouped Bar Chart")
plt.show()
```

#4. Stacked Bar Chart
Definition:
Shows parts of a whole — bars are stacked on top of each other (vertical or horizontal).

Used for:
Showing composition of categories over time or across groups.

```python
plt.bar(categories, values_2023, label='2023')
plt.bar(categories, values_2024, bottom=values_2023, label='2024')
plt.legend()
plt.title("Stacked Bar Chart")
plt.show()
```

# 100% Stacked Bar Chart
Definition:
Like a stacked bar chart, but each bar’s total = 100%, showing proportions instead of raw values.

Used for:
Comparing percentage composition among categories.

```python
categories = ['A', 'B', 'C']
male = np.array([30, 40, 20])
female = np.array([70, 60, 80])

# Convert to percentages
total = male + female
male_percent = male / total * 100
female_percent = female / total * 100

plt.bar(categories, male_percent, label='Male')
plt.bar(categories, female_percent, bottom=male_percent, label='Female')

plt.ylabel('Percentage (%)')
plt.title('100% Stacked Bar Chart')
plt.legend()
plt.show()
```

#6. Diverging Bar Chart
Definition:
Bars extend both left and right from a central axis (often used for positive/negative comparisons).

Used for:
Displaying differences or sentiments (e.g., agree vs. disagree, profit vs. loss)

## Vertical diverging bar

```python
import matplotlib.pyplot as plt

categories = ['Product A', 'Product B', 'Product C', 'Product D']
values = [25, -10, 15, -5]

colors = ['green' if v >= 0 else 'red' for v in values]

plt.bar(categories, values, color=colors)
plt.axhline(0, color='black', linewidth=1)  # baseline at 0

plt.title('Diverging Bar Chart (Positive vs Negative)')
plt.ylabel('Change')
plt.show()
```

## Horizzontal Diverging Bar

```python
categories = ['A', 'B', 'C', 'D']
values = [20, -15, 30, -10]

colors = ['green' if v >= 0 else 'red' for v in values]

plt.barh(categories, values, color=colors)
plt.axvline(0, color='black', linewidth=1)  # central zero line

plt.title('Diverging Bar Chart')
plt.xlabel('Value')
plt.show()
```

# 7. Floating Bar Chart
Definition:
Each bar starts and ends at a different value (not from zero).

Used for:
Showing ranges (e.g., temperature range, price range).

```python
categories = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
low = np.array([10, 12, 14, 11, 13])
high = np.array([20, 22, 25, 21, 23])

# height = difference between high and low
plt.bar(categories, high - low, bottom=low, color='skyblue')

plt.title('Floating Bar Chart (Temperature Range)')
plt.ylabel('Temperature (°C)')
plt.show()
```

# 8. Stacked Grouped Bar Chart (Hybrid)
Definition:
Combines grouping and stacking — each group of bars is stacked within a category.

Used for:
Detailed comparison across multiple categories and time frames.

```python
categories = ['Q1', 'Q2', 'Q3']
x = np.arange(len(categories))
width = 0.35

# Data for two years (each with two stacked parts)
profit_2023 = np.array([15, 20, 25])
loss_2023 = np.array([5, 7, 4])
profit_2024 = np.array([18, 22, 30])
loss_2024 = np.array([6, 5, 6])

# Group 1: 2023
plt.bar(x - width/2, profit_2023, width, label='Profit 2023')
plt.bar(x - width/2, loss_2023, width, bottom=profit_2023, label='Loss 2023')

# Group 2: 2024
plt.bar(x + width/2, profit_2024, width, label='Profit 2024')
plt.bar(x + width/2, loss_2024, width, bottom=profit_2024, label='Loss 2024')

plt.xticks(x, categories)
plt.title('Stacked + Grouped Bar Chart')
plt.legend()
plt.show()
```
